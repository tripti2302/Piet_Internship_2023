```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import time
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree #for creating tree in visual form
```

```python
iris=load_iris()
iris
iris.target_names
iris.feature_names
features=iris.data
labels=iris.target
iris.target_names
iris.feature_names
features
```

```python
features1=features[:50,:]
features1
```

```python
x=features1[:,0]
y=features1[:,1]
x1=features1[:,2]
y1=features1[:,3]
plt.xlabel('sepal_petal_length')
plt.ylabel('sepal_petal_width')
plt.title('setosa')
plt.scatter(x,y,color='red')
plt.scatter(x1,y1,color='green',marker='*')
plt.grid()
plt.show()
```

```python
#lets split data
#python code for sklearn
trainX,testx,trainY,testy=train_test_split(features,labels,test_size=0.2)
model=DecisionTreeClassifier(criterion='entropy')
#training model 
trained_model=model.fit(trainX,trainY)
predicted_output=trained_model.predict(testx)
print(predicted_output)


```

```python
testy
```

```python
score=accuracy_score(testy,predicted_output)
print("accuracy of decision tree is",score)
```

```python
#lets plot tree structure
plt.figure(figsize=(12,8))
plot_tree(trained_model,feature_names=iris.feature_names,class_names=iris.target_names.tolist(),filled=True)
plt.title('decision tree')
plt.show()
```
